```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 3 
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
```
