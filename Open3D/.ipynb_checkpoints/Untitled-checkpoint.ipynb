```python
from tkinter import*
import open3d as o3d
import pandas as pd
import numpy as np

global list_of_selected_points_1
list_of_selected_points_1=[]

class trial_class_in_tkinter:
   
    def __init__(self,master,colors=[],colors_list=[],description="",list_of_selected_points_1=[],
                 description_list = [],dictionary_for_line={},list_of_len_of_points=[],
                 color_list_1 = [], color_list_2 = [], color_list_3 = [], color_list_4 = [],
                 dictionary = {}, list_of_selected_points=[],colors_mesh=[],colors_list_mesh=[],description_mesh="",
                 description_list_mesh = [], color_list_1_mesh = [], color_list_2_mesh = [], color_list_3_mesh = [], color_list_4_mesh = [],
                 dictionary_for_mesh = {}, list_of_selected_points_for_mesh=[],
                 appended_list_of_selected_points_for_mesh=[], mesh_append_result = []):
        frame = Frame(master)
        frame.grid()
        self.colors = colors
        self.list_of_selected_points_1 = list_of_selected_points_1
        self.description = description
        self.dictionary_for_line = dictionary_for_line
        self.list_of_len_of_points = list_of_len_of_points
        self.description_list = description_list
        self.colors_list = colors_list
        self.colors_list_1 = color_list_1
        self.colors_list_2 = color_list_2
        self.colors_list_3 = color_list_3
        self.colors_list_4 = color_list_4
        self.dictionary = dictionary
        self.dictionary_for_line = dictionary_for_line
        self.list_of_selected_points = list_of_selected_points
        self.list_of_len_of_points = list_of_len_of_points
        self.colors_mesh = colors_mesh
        self.list_of_selected_points_1 = list_of_selected_points_1
        self.description_mesh = description_mesh
        self.dictionary_for_line = dictionary_for_line
        self.list_of_len_of_points = list_of_len_of_points
        self.description_list_mesh = description_list_mesh
        self.colors_list_mesh = colors_list_mesh
        self.colors_list_1_mesh = color_list_1_mesh
        self.colors_list_2_mesh = color_list_2_mesh
        self.colors_list_3_mesh = color_list_3_mesh
        self.colors_list_4_mesh = color_list_4_mesh
        self.dictionary = dictionary
        self.dictionary_for_mesh = dictionary_for_mesh
        self.list_of_selected_points_for_mesh = list_of_selected_points_for_mesh
        self.appended_list_of_selected_points_for_mesh = appended_list_of_selected_points_for_mesh
        self.mesh_append_result = mesh_append_result
       
        self.file_label = Label(frame,text="Enter the file path here")
        self.file_label.grid(row=0,column=0)
        self.entry_file = Entry(frame, width = 70)
        self.entry_file.grid(row=0,column=1)
       
        self.color_label_mesh = Label(frame,text="Enter the rgb code here")
        self.color_label_mesh.grid(row=14,column=0)
        self.entry_color_list_1_mesh = Entry(frame, width = 30)
        self.entry_color_list_1_mesh.grid(row=15,column=0)

        self.entry_color_list_2_mesh = Entry(frame, width = 30)
        self.entry_color_list_2_mesh.grid(row=16,column=0)
       
        self.entry_color_list_3_mesh = Entry(frame, width = 30)
        self.entry_color_list_3_mesh.grid(row=17,column=0)
       
        self.entry_color_list_4_mesh = Entry(frame, width = 30)
        self.entry_color_list_4_mesh.grid(row=18,column=0)
       
        self.description_label_mesh = Label(frame,text="Enter the description here")
        self.description_label_mesh.grid(row=14,column=1)
        self.description_color_list_1_mesh = Entry(frame, width = 30)
        self.description_color_list_1_mesh.grid(row=15,column=1)

        self.description_color_list_2_mesh = Entry(frame, width = 30)
        self.description_color_list_2_mesh.grid(row=16,column=1)
       
        self.description_color_list_3_mesh = Entry(frame, width = 30)
        self.description_color_list_3_mesh.grid(row=17,column=1)
       
        self.description_color_list_4_mesh = Entry(frame, width = 30)
        self.description_color_list_4_mesh.grid(row=18,column=1)
       
        self.printButton_3 = Button(frame, text="Create mesh", command = self.mesh_creation,width=25,padx=10,pady=10)
        self.printButton_3.grid(row=20,column=1)
       
        self.printButton_2 = Button(frame, text="Start calculation[mesh]", command = self.loopint_function_mesh,width=25,padx=10,pady=10)
        self.printButton_2.grid(row=19,column=0)
       
        self.file_label = Label(frame,text="Enter the file path here")
        self.file_label.grid(row=0,column=0)
        self.entry_file = Entry(frame, width = 70)
        self.entry_file.grid(row=0,column=1)
       
        self.color_label = Label(frame,text="Enter the rgb code here")
        self.color_label.grid(row=3,column=0)
        self.entry_color_list_1 = Entry(frame, width = 30)
        self.entry_color_list_1.grid(row=4,column=0)

        self.entry_color_list_2 = Entry(frame, width = 30)
        self.entry_color_list_2.grid(row=5,column=0)
       
        self.entry_color_list_3 = Entry(frame, width = 30)
        self.entry_color_list_3.grid(row=6,column=0)
       
        self.entry_color_list_4 = Entry(frame, width = 30)
        self.entry_color_list_4.grid(row=7,column=0)
       
        self.description_label = Label(frame,text="Enter the description here")
        self.description_label.grid(row=3,column=1)
        self.description_color_list_1 = Entry(frame, width = 30)
        self.description_color_list_1.grid(row=4,column=1)

        self.description_color_list_2 = Entry(frame, width = 30)
        self.description_color_list_2.grid(row=5,column=1)
       
        self.description_color_list_3 = Entry(frame, width = 30)
        self.description_color_list_3.grid(row=6,column=1)
       
        self.description_color_list_4 = Entry(frame, width = 30)
        self.description_color_list_4.grid(row=7,column=1)
       
#        self.printButton = Button(frame, text="Get points", command = self.get_points_function,width=25,padx=10,pady=10)
#        self.printButton.grid(row=2,column=1)
       
        self.printButton_1 = Button(frame, text="Start calculation [Lines]", command = self.loopint_function_lines,width=25,padx=10,pady=10)
        self.printButton_1.grid(row=12,column=0)
#        m = 0
#        while m <3:
#            arr = (self.entry_color_list_1.get()).split(" ")
#            arr_int=[]
#            for i in range(len(arr)):
#                arr_int.append(arr[i])
#        print(arr_int)
#        self.colors_list = [self.entry_color_list_1.get(),self.entry_color_list_2.get(),self.entry_color_list_3(),self.entry_color_list_4()]
       

#        
    def loopint_function_lines(self):
        m = 0
        self.get_dictionary()
        while m <4:
            self.pick_points()
#            arr = (self.entry_color_list_1.get()).split(" ")
            if self.entry_color_list_1.get() == "yellow":
                self.colors_list_1= [1, 1, 0]
            elif self.entry_color_list_1.get() == "red":
                self.colors_list_1= [1, 0, 0]
            if self.entry_color_list_2.get() == "yellow":
                self.colors_list_2= [1, 1, 0]
            elif self.entry_color_list_2.get() == "red":
                self.colors_list_2= [1, 0, 0]
            if self.entry_color_list_3.get() == "yellow":
                self.colors_list_3= [1, 1, 0]
            elif self.entry_color_list_3.get() == "red":
                self.colors_list_3= [1, 0, 0]
            if self.entry_color_list_4.get() == "yellow":
                self.colors_list_4= [1, 1, 0]
            elif self.entry_color_list_4.get() == "red":
                self.colors_list_4= [1, 0, 0]

            self.colors_list = [self.colors_list_1]+[self.colors_list_2]+[self.colors_list_3]+[self.colors_list_4]
#            print(self.colors_list)
            self.colors = self.colors_list[m]

#            self.description = input("Enter this polyline description: ")
            self.description_list = [self.description_color_list_1.get()]+[self.description_color_list_2.get()]+[self.description_color_list_3.get()]+[self.description_color_list_4.get()]
            self.description = self.description_list[m]
            print(self.description)
#            for i in range(len(self.description_list)):
#                if self.description_list[i] == self.description:
#                    self.colors = self.colors_list[i]
            self.points_to_line_calculation_function()
            self.convert_array_to_ply(m)
            m = m +1
        self.visualization_of_polylines_with_tunnel()
       
#    def get_points_function(self):
#        m = 0
#        self.get_dictionary()
#        while m<4:
#            self.pick_points()
#            m = m + 1
    def loopint_function_mesh(self):
        m = 0
        self.get_dictionary()
        while m <4:
            self.pick_points_for_mesh()
            if self.entry_color_list_1_mesh.get() == "yellow":
                self.colors_list_1_mesh= [1, 1, 0]
            elif self.entry_color_list_1_mesh.get() == "red":
                self.colors_list_1_mesh= [1, 0, 0]
            if self.entry_color_list_2_mesh.get() == "yellow":
                self.colors_list_2_mesh= [1, 1, 0]
            elif self.entry_color_list_2_mesh.get() == "red":
                self.colors_list_2_mesh= [1, 0, 0]
            if self.entry_color_list_3_mesh.get() == "yellow":
                self.colors_list_3_mesh= [1, 1, 0]
            elif self.entry_color_list_3_mesh.get() == "red":
                self.colors_list_3_mesh= [1, 0, 0]
            if self.entry_color_list_4_mesh.get() == "yellow":
                self.colors_list_4_mesh= [1, 1, 0]
            elif self.entry_color_list_4_mesh.get() == "red":
                self.colors_list_4_mesh= [1, 0, 0]
            self.colors_list_mesh = [self.colors_list_1_mesh]+[self.colors_list_2_mesh]+[self.colors_list_3_mesh]+[self.colors_list_4_mesh]
#            print(self.colors_list)
            self.colors_mesh = self.colors_list_mesh[m]
            self.description_list_mesh = [self.description_color_list_1_mesh.get()]+[self.description_color_list_2_mesh.get()]+[self.description_color_list_3_mesh.get()]+[self.description_color_list_4_mesh.get()]
            self.description_mesh = self.description_list_mesh[m]
            print(self.description_mesh)
            self.convert_array_to_mesh(m)
            m = m +1
       
    def mesh_creation(self):
        self.create_mesh()
        self.visualization_of_mesh_with_tunnel()


    def get_dictionary(self):
        file = "ply_1.ply"
#        file = self.entry_file.get()
#        file = "C:\\Users\\Arjun Janamatti\\Downloads\\Tunnel.ply"

        pcd_load = o3d.io.read_point_cloud("ply_1.ply")
        xyz_load = np.asarray(pcd_load.points)
        rgb_load = np.asarray(pcd_load.colors)
        data = np.concatenate((xyz_load,rgb_load), axis=1)
        dataframe = pd.DataFrame(data=data[0:,0:],index=[i for i in range(data.shape[0])],columns=['f'+str(i) for i in range(data.shape[1])])
        self.dictionary = dataframe.T.to_dict("list")
        for i in range(len(self.dictionary)):
            self.dictionary[i] = self.dictionary[i][:3]
#        print("")
#        print(
#            "1) Please pick at least three correspondences using [shift + left click]"
#        )
#        print("   Press [shift + right click] to undo point picking")
#        print("2) After picking points, press q for close the window")
    def pick_points(self):
        self.list_of_selected_points_1 = []
        file = "ply_1.ply"
        pcd = o3d.io.read_point_cloud(file)
        vis = o3d.visualization.VisualizerWithEditing()
        vis.create_window()
        vis.add_geometry(pcd)
        vis.run()  # user picks points
        vis.destroy_window()
        point_number_in_form_of_list = vis.get_picked_points()
        for i in point_number_in_form_of_list:
            print(i)
            for j in self.dictionary:
                if i == j:
                    self.list_of_selected_points_1.append(self.dictionary[i])
       
        print(self.list_of_selected_points_1)
        self.list_of_len_of_points.append(len(self.list_of_selected_points_1))

    def pick_points_for_mesh(self):
        self.list_of_selected_points_for_mesh = []
        file = "ply_1.ply"
        pcd = o3d.io.read_point_cloud(file)
        vis = o3d.visualization.VisualizerWithEditing()
        vis.create_window()
        vis.add_geometry(pcd)
        vis.run()  # user picks points
        vis.destroy_window()
        point_number_in_form_of_list = vis.get_picked_points()
        for i in point_number_in_form_of_list:
            print(i)
            for j in self.dictionary:
                if i == j:
                    self.list_of_selected_points_for_mesh.append(self.dictionary[i])
       
        self.appended_list_of_selected_points_for_mesh.append(self.list_of_selected_points_for_mesh)
        print(len(self.appended_list_of_selected_points_for_mesh))
                   
    def points_to_line_calculation_function(self):
       
#        print("Inside second definition")
        print(self.list_of_selected_points_1)
        for i in range(len(self.list_of_selected_points_1)):
            if i < len(self.list_of_selected_points_1) - 1:
                C_1,D_1,E_1 = self.list_of_selected_points_1[i]
                C_2,D_2,E_2 = self.list_of_selected_points_1[i+1]
                F_1 = E_2-E_1
                I_1 = 0.01
                G_2 = np.sqrt((C_2-C_1)**2 + (D_2-D_1)**2)
                H_1 = I_1 * F_1/G_2
                print("inside for loop in second devision")
       
       
                H_2 = np.where((C_2==C_1) & (D_2==D_1),
                               np.arctan((0.0001+C_1-C_2)/(0.0001+D_1-D_2))*(180/np.pi),
                               (np.arctan((C_1-C_2)/(D_1-D_2))*(180/np.pi)))
       
                I_2 = C_2-C_1
                J_2 = D_2-D_1
                K_2 = np.where((I_2>0)&(J_2>0),H_2*1,0)
                L_2 = np.where((I_2>0)&(J_2<0),H_2+180,0)
                M_2 = np.where((I_2<0)&(J_2<0),H_2+180,0)
                N_2 = np.where((I_2<0)&(J_2>0),H_2+360,0)
                F_2 = np.sum(np.array([K_2,L_2,M_2,N_2]))
                points_for_x = []
                points_for_y = []
                points_for_z = []
                c_first = C_1
                c_last = C_2
                d_first = D_1
                d_last = D_2
                e_first = E_1
                e_last = E_2
                while abs(c_last - c_first) >= 0.01:
                    c_first = c_first + (I_1 * (np.sin(F_2*np.pi/180)))
                    d_first = d_first + (I_1 * (np.cos(F_2*np.pi/180)))
                    e_first = e_first + H_1
                    points_for_x.append(c_first)
                    points_for_y.append(d_first)
                    points_for_z.append(e_first)
                self.dictionary_for_line["line_{0}".format(i)] = [points_for_x,points_for_y,points_for_z]
#            print(self.dictionary_for_line.keys())
           
           
    def convert_array_to_ply(self,m):
        pcd = o3d.geometry.PointCloud()
#        for i in range(len(self.description_list)):
#            if self.description_list[i] == self.description:
#                self.colors = self.colors_list[i]
#                break
#                print("this is in for loop for description: ", self.colors)
        for i in range(len(self.list_of_selected_points_1)):
            for j in range(len(self.list_of_selected_points_1)-1):
                print(self.colors)
                pcd.points = o3d.utility.Vector3dVector(np.transpose(np.array(self.dictionary_for_line["line_{0}".format(j)])))
                pcd.paint_uniform_color(self.colors)
                o3d.io.write_point_cloud("line_{0}".format(m)+"{0}.ply".format(j), pcd)

    def convert_array_to_mesh(self,m):
        pcd = o3d.geometry.PointCloud()
        for i in range(len(self.appended_list_of_selected_points_for_mesh)):
#            for j in range(len(self.list_of_selected_points_for_mesh)-1):
            pcd = o3d.geometry.PointCloud()
            pcd.points = o3d.utility.Vector3dVector(np.array(self.appended_list_of_selected_points_for_mesh[m]))                
            pcd.paint_uniform_color(self.colors_mesh)
            o3d.io.write_point_cloud("mesh_{0}".format(m)+".ply", pcd)
           
    def create_mesh(self):
        for i in range(len(self.appended_list_of_selected_points_for_mesh)):
            pcd = o3d.io.read_point_cloud("mesh_{0}".format(i)+".ply")
            pcd.estimate_normals()
           
            # estimate radius for rolling ball
            distances = pcd.compute_nearest_neighbor_distance()
            avg_dist = np.mean(distances)
            radius = 1.5 * avg_dist  
           
            mesh = o3d.geometry.TriangleMesh.create_from_point_cloud_ball_pivoting(
                       pcd,
                       o3d.utility.DoubleVector([radius, radius * 2]))
            self.mesh_append_result.append(mesh)

    def visualization_of_polylines_with_tunnel(self):
        print(self.list_of_len_of_points)
        vis = o3d.visualization.Visualizer()
        vis.create_window()
        for n in range(len(self.list_of_len_of_points)):
            for j in range(self.list_of_len_of_points[n]-1):    
#                vis.add_geometry(o3d.io.read_point_cloud("C:\\Users\\Arjun Janamatti\\Downloads\\Tunnel.ply"))
#                "C:\\Users\\Arjun Janamatti\\Downloads\\ply_1.ply"
                vis.add_geometry(o3d.io.read_point_cloud("ply_1.ply"))
                vis.add_geometry(o3d.io.read_point_cloud("line_{0}".format(n)+"{0}.ply".format(j)))
        vis.run()  # user picks points
        vis.destroy_window()

    def visualization_of_mesh_with_tunnel(self):
#         print(self.list_of_len_of_points_for_mesh)
        vis = o3d.visualization.Visualizer()
        vis.create_window()
        for n in range(len(self.appended_list_of_selected_points_for_mesh)):
#            for j in range(self.list_of_len_of_points_for_mesh[n]-1):    
#                vis.add_geometry(o3d.io.read_point_cloud("C:\\Users\\Arjun Janamatti\\Downloads\\Tunnel.ply"))
#                "C:\\Users\\Arjun Janamatti\\Downloads\\ply_1.ply"
            vis.add_geometry(o3d.io.read_point_cloud("ply_1.ply"))
            vis.add_geometry(self.mesh_append_result[n])
        vis.run()  # user picks points
        vis.destroy_window()            
   
   

root = Tk()
root.geometry("650x300")
b = trial_class_in_tkinter(root)
root.mainloop()
```
