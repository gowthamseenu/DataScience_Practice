```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional method
```

```python
plt.plot(x,y)
#plt.show() for other IDE and it's like printing graph
```

```python
plt.plot(x,y,'-r')
```

```python
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'y')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
#Object Oriented method
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.45,0.4,0.35])
axes1.plot(x,y)
axes1.set_title('Larger Plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```
